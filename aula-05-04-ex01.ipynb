```python
import pandas as pd
import numpy as np
url = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
df = pd.read_csv(url, encoding='latin-1', sep=';')
df.head()
```

```python
# Agrupando por CISP e somando os furtos de celular
df_cisp = df.groupby('cisp')['furto_celular'].sum().reset_index()
# Ordenando do maior para o menor
df_cisp = df_cisp.sort_values(by='furto_celular', ascending=False)
```
